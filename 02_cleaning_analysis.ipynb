###1. Imports iniciales

```python
# Librerías principales
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Configuración general
plt.rcParams['figure.figsize'] = (10,5)
plt.style.use('default')
```

###2. Cargar datos

```python
df = pd.read_csv('/content/market_data.csv', skiprows=2)

df.columns = [
    "Date",
    "MXN_Close", "GSPC_Close", "IXIC_Close", "MXX_Close",
    "MXN_High",  "GSPC_High",  "IXIC_High",  "MXX_High",
    "MXN_Low",   "GSPC_Low",   "IXIC_Low",   "MXX_Low",
    "MXN_Open",  "GSPC_Open",  "IXIC_Open",  "MXX_Open",
    "MXN_Volume","GSPC_Volume","IXIC_Volume","MXX_Volume"
]

df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

###3. Información del dataset

```python
df.info()
```

###4. Descripción estadística

```python
df.describe().T
```

###5. Verificar valores faltantes

```python
df.isnull().sum()
```

###6. Manejo de valores faltantes

```python
# Eliminar filas con NA
df_clean = df.dropna()
```

###7. Valores duplicados

```python
df_clean.duplicated().sum()
df_clean = df_clean.drop_duplicates()
```

###8. Conversión de tipos

```python
# Convertir fecha
if 'date' in df_clean.columns:
    df_clean['date'] = pd.to_datetime(df_clean['date'])

# Asegurar numéricos
cols_num = ['open', 'close', 'high', 'low', 'volume']
for c in cols_num:
    if c in df_clean.columns:
        df_clean[c] = pd.to_numeric(df_clean[c], errors='coerce')
```

###9. Análisis Exploratorio

####9.1 Gráfico de línea

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["MXN_Close"], label="USD/MXN")
plt.plot(df["Date"], df["GSPC_Close"], label="S&P500")
plt.plot(df["Date"], df["IXIC_Close"], label="NASDAQ")
plt.plot(df["Date"], df["MXX_Close"], label="IPC México")

plt.title("Tendencias históricas de los mercados")
plt.xlabel("Fecha")
plt.ylabel("Valor de cierre")
plt.legend()
plt.grid(True)
plt.show()
```

####9.2 Distribución de precios

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.boxplot(data=df[["MXN_Close", "GSPC_Close", "IXIC_Close", "MXX_Close"]])
plt.title("Distribución de precios de cierre")
plt.show()
```

####9.3 Matriz de correlación

```python
corr = df[["MXN_Close", "GSPC_Close", "IXIC_Close", "MXX_Close"]].corr()
corr
```

####9.4 Heatmap de correlación

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Heatmap de correlación entre Índices Financieros")
plt.show()
```
